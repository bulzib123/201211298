##### regex

```python
import urllib
uResponse = urllib.urlopen('http://python.org/')
_html=uResponse.read()
```

```python
import re
p=re.compile('href="(http://.*?)"')
nodes=p.findall(_html)
for i, node in enumerate(nodes):
    print i+1,node
```

##### BeautifulSoup

```python
from urllib import urlopen
from bs4 import BeautifulSoup
_html = urlopen('http://python.org/').read()
tree = BeautifulSoup(_html, "lxml")
for link in tree.findAll("a"):
    if 'href' in link.attrs:
        print link.attrs['href']
```

```python
from urllib import urlopen
from bs4 import BeautifulSoup
import requests
page = requests.get('https://www.python.org/')
soup = BeautifulSoup(page.content)
for node in soup.find_all("a"):
    if 'href' in node.attrs:
        print node
```

##### xpath

```python
from lxml import etree
_htmlTree = etree.HTML(_html)

nodes = _htmlTree.xpath('//*[@href]')

for node in nodes:
    print node.attrib
```

##### css selector

```python
import lxml.html
from lxml.cssselect import CSSSelector
import requests
r = requests.get('http://python.org/')

html=lxml.html.fromstring(r.text)
sel=CSSSelector('a[href]')
nodes = sel(html)

for node in nodes:
    print lxml.html.tostring(node)
```

##### 국제 학회 목록 크롤링

```python
%%writefile src/ds_web_crawl_ieee2.py
# coding: utf-8
import lxml.html
from lxml.cssselect import CSSSelector
import requests
r = requests.get('http://www.ieee.org/conferences_events/conferences/search/index.html')
html = lxml.html.fromstring(r.text)

sel = CSSSelector('div.content-r-full table.nogrid-nopad tr p>a[href]')
nodes = sel(html)

for node in nodes:
    print node.text
    print lxml.html.tostring(node)
    print"--"
```
